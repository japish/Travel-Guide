```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
travel_data=pd.read_csv('Top_Indian_Places_to_Visit.csv')
```

```python
# Convert the 'State' column to categorical type
travel_data['State'] = travel_data['State'].astype('category')

# Print the information of the dataframe
travel_data.info()
```

```python
#Which is the state with most visiting places?
fig,ax=plt.subplots(figsize=(20,12))
sns.countplot(y='State',data=travel_data)
```

```python
#Which are the top 5 google rated visiting place types
top_5_names=travel_data['Significance'].value_counts().head(5).index
top_5_rating_place=travel_data[travel_data['Significance'].isin(top_5_names)]

plt.figure(figsize=(20,10))
sns.countplot(x='Google review rating',hue='Significance',data=top_5_rating_place)
plt.title('Top 5 Rated Place Types')
plt.show()
```

```python
#What is the time required to visit different types of places
plt.figure(figsize=(10,7))
sns.boxplot(x='time needed to visit in hrs',y='Significance',data=travel_data,palette='Set3')
plt.title('Distribution of Time Needed to Visit By Significance')
plt.xlabel('Significance')
plt.ylabel('Time Needed to Visit (hours)')
plt.grid(True)
plt.show()
```

```python
#What are the best time to visit places?
top_5_city = travel_data['Name'].head(20).values
top_5_city_data = travel_data[travel_data['Name'].isin(top_5_city)]
plt.figure(figsize=(10, 5))
sns.countplot(y='Name', hue='Best Time to visit', data=top_5_city_data)
plt.title('Places with Best Time to Visit')
plt.show()
```

```python
# What are the types of visiting places that permits DSLR?
sns.countplot(y='Significance',hue='DSLR Allowed',data=travel_data)
```

```python
#Which state has the most google ratings?
plt.figure(figsize=(25, 6))
sns.barplot(x='State', y='Google review rating', data=travel_data, errorbar=None, palette='viridis')
plt.title('Average Google Review Rating by State')
plt.xlabel('State')
plt.ylabel('Average Google Review Rating in lakhs')
plt.xticks(rotation=45)
plt.show()
```

```python
#What are the top 10 establishment types?
top_types = travel_data['Type'].value_counts().nlargest(10).index
plt.figure(figsize=(12, 6))
sns.countplot(x='Type', data=travel_data[travel_data['Type'].isin(top_types)], order=top_types, palette='Set2')
plt.title('Top Ten Establishments by Type')
plt.xlabel('Establishment Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#What zone has the highest percentage of places to visit in India?
type_distribution = travel_data['Zone'].value_counts()

plt.figure(figsize=(10, 6))
type_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['gold', 'lightgreen', 'lightcoral', 'lightskyblue'])
plt.title('Distribution of Types of Places')
plt.ylabel('')
plt.show()
```

```python
#How much time needed to visit different places vs google review rating
plt.figure(figsize=(10, 6))
plt.scatter(travel_data['time needed to visit in hrs'], travel_data['Google review rating'], color='red', alpha=0.7)
plt.title('Time Needed to Visit vs Google Review Rating')
plt.xlabel('Time Needed to Visit (hrs)')
plt.ylabel('Google Review Rating')
plt.grid()
plt.show()
```

```python
#show number of places by type*such as historical, natural, etc,) in each state of india
plt.figure(figsize=(18, 6))
state_type_count = travel_data.groupby(['State', 'Significance']).size().unstack().fillna(0)
state_type_count.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Number of Places by Type in Each State')
plt.xlabel('State')
plt.ylabel('Number of Places')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))  # Adjusting the legend position
plt.show()
```

```python
#Create scatterplot graph of the number of google reviews in lacs vs google review rating for different types of places
sns.scatterplot(x='Number of google review in lakhs', y='Google review rating', hue='Significance', data=travel_data)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.subplots_adjust(right=0.75)
plt.show()
```

```python
sns.histplot(travel_data['Google review rating'], kde=True)
plt.show()
```

```python
# 6. Heatmap

corr = travel_data.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
